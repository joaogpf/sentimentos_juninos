```python

nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()


#tweets = pd.read_csv('tweets.csv')

#pré-processamento dos dados
def preprocessamento(texto):
    # Remove emotes
    emote_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    texto = emote_pattern.sub(r'', texto)
    texto = re.sub(r'http\S+', '', texto)
    texto = re.sub(r'@\w+', '', texto)
    texto = re.sub(r'#', '', texto)
    texto = re.sub(r'[^\w\s]', '', texto)

    # Remove stopwords
    stop_words = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = texto.lower().split()
    palavras_sem_stopwords = [palavra for palavra in palavras if palavra not in stop_words]
    texto_sem_stopwords = " ".join(palavras_sem_stopwords)

    #lematização
    tokens = word_tokenize(texto_sem_stopwords)
    print(tokens)
    texto_lematizado = ' '.join([lemmatizer.lemmatize(token) for token in tokens])

    return texto_lematizado

#df_2023['tweet'] = df_2023['tweet'].astype(str)
#df_2024['post'] = df_2024['post'].apply(preprocessamento)
#tweets['post'] = tweets['post'].apply(preprocessamento)
#tweets['sentimento'] = tweets['post'].apply(classificar_frase)
#tweets.to_csv('tweets.csv', index=False)
#df_2024.to_csv('tweets_2024_preprocessados.csv', index=False)
#df_Noticias['corpus'] = df_Noticias['corpus'].apply(preprocessamento)
#df_Noticias.to_csv('noticias_preprocessadas.csv', index=False)

```

Classificando Sentimentos com VADER



```python
#df_tweets_ingles = pd.read_csv('/content/tweets_2024_traduzidos.csv')
#df_tweets_2024 = pd.read_csv('/content/tweets_2024_preprocessados.csv')

df_noticias_preprocessadas = pd.read_csv('/content/noticias_preprocessadas.csv')
df_noticias_preprocessadas.head()
analisador = SentimentIntensityAnalyzer()

def analisar_sentimento(texto):
    pontuacao = analisador.polarity_scores(texto)
    if pontuacao['compound'] >= 0.05:
        return 'positiva'
    elif pontuacao['compound'] <= -0.05:
        return 'negativa'
    else:
        return 'neutra'


df_noticias_preprocessadas['sentimento'] = df_noticias_preprocessadas['corpus'].apply(analisar_sentimento)
df_noticias_preprocessadas.to_csv('noticias_classificados.csv', index=False)

sentimentos = pd.read_csv('/content/noticias_classificados.csv')
sentimentos
```
